# MAPE

```python
# true_vector 와 predict_vector 를 가지고 오차 vector을 추출한다  
def mae_error(true, predicted):
    res = np.empty(true.shape)
    for j in range(true.shape[0]):
        if true[j] != 0:
            #MAE 계산 
            res[j] = (true[j] - predicted[j]) / true[j]
        else:
            #실제 값이 0 인 경우 나누는 값이 0 이되므로 문제가 생겨 이러한 상황에는 실제값 전체 벡터의 평균으로 나누어준다.
            #이러한 방법이 맞는지 논의하기 
            res[j] = predicted[j] / np.mean(true)
    return np.abs(res)

#오차 vector 의 total MAPE를 계산한다. 
def mean_absolute_percentage_error(y_true, y_pred): 
    error_vector=mae_error(np.asarray(y_true), np.asarray(y_pred))
    MAPE = np.mean(error_vector) * 100
    # MAPE 계산해서 return
    return error_vector, MAPE
```

```python
#error 값 계산하는 부분 
error_df = pd.DataFrame(columns=prediction.columns)
MAPE_list=[]
for j in range(len(prediction)):
    # row 한개씩 가져오기 
    # return : error_vector 와 MAPE 
    error,MAPE=mean_absolute_percentage_error(true.iloc[j],prediction.iloc[j])
    MAPE_list.append(MAPE)
    error_df.loc[j]=error.T
    
```

```python
# error_df 에 mape column추가
error_df["MAPE"]=MAPE_list
```

```python
# MAPE는 MAE를 percentage로 나타낸값이므로 
# 오차율을 mean 처리함으로써 MAE값으로 바꾸고 그 값 * 100을하여 percentage 로 나타낸다. 
# 세로 column MAPE 계산하는 함수 
def mape_making(error): 
    return np.mean(error) * 100
```

```python
error_df=error_df.drop("MAPE",axis=1)
```

```python
mape_vector = []
for i in range(len(error_df.columns)):
    value=error_df.iloc[:,i].values
    mape=mape_making(value)
    mape_vector.append(mape)
    #print("column = ",i,"and MAPE = ",mape)
```

```python
MAPE_vector=pd.DataFrame(mape_vector).T
```

# RMSE

```python
def rmse(predictions, true):

    differences = predictions - true                       #the DIFFERENCEs.

    differences_squared = differences ** 2                    #the SQUAREs of 

    mean_of_differences_squared = differences_squared.mean()  #the MEAN of 

    rmse_val = np.sqrt(mean_of_differences_squared)           #ROOT of 

    return rmse_val                                           
```

```python
# true_vector 와 predict_vector 를 가지고 오차 vector을 추출한다  
# rmse 방법에서의 오차는 (prediction-true)**2 로 계산한다. 
def rmse_error(true, pred):
    res = np.empty(true.shape)
    for j in range(true.shape[0]):
        res[j]= (pred[j]-true[j])**2
    return res

#오차 vector 의 total MAPE를 계산한다. 
def root_mean_squared_error(y_true, y_pred): 
    error_vector=rmse_error(np.asarray(y_true), np.asarray(y_pred))
    RMSE = np.sqrt(np.mean(error_vector))
    return error_vector, RMSE
```

```python
#error 값 계산하는 부분 
error_dfForRmse = pd.DataFrame(columns=prediction.columns)
RMSE_list=[]
for j in range(len(prediction)):
    # row 한개씩 가져오기 
    # return : error_vector 와 MAPE 
    error,RMSE=root_mean_squared_error(true.iloc[j],prediction.iloc[j])
    RMSE_list.append(RMSE)
    error_dfForRmse.loc[j]=error.T
    
```

```python
# error_df 에 rmse column추가
error_dfForRmse["RMSE"]=RMSE_list
```
